# 1. Descripción del problema

El problema a tratar por este algoritmo es el **problema de ordenamiento**, problema que se ha estudiado por más de 50 años y consiste en modificar una secuencia de números para que luego de aplicado el algoritmo, los mismos elementos de la secuencia se encuentren en un orden específico. Por ejemplo ordenar los números de un arreglo de menor a mayor o viceversa.

**Entrada:** Secuencia de *n* números $[a_1,a_2,...,a_n]$

**Salida**: Permutación ordenada de la secuencia de entrada: $[a_1',a_2',...,a_n']$, de tal manera que $a_1'\leq a_2' \leq... \leq a_n'$.

**Tamaño del problema:** El tamaño para este problema viende definido por la cantidad *n* de valores a ordenar.

# 2. Insertion sort


## 2.1 Código

```python
def insertionSort(array, verbose = False): #Argumentos: El arreglo a ordenar + opción verbose para ver el paso a paso
  if verbose == True:
    print("\n")

  comp = 0 #comparaciones
  permut = 0 #permutaciones

  for i in range (1,len(array)): #Ciclo más grande, es el que realiza cada paso del algoritmo
    key = array[i] #Se guarda número a ubicar
    j = i-1

    if verbose == True:
      print("Pasada", i)
      print("arreglo al inicio:", array)
      print("Número a posicionar:", array[i])

    while 1:
      comp += 1
      if j >= 0 and key < array[j]:
        permut+=1
        #Si el número es mayor al de su izquierda, se intercambian de posición entre ellos
        array[j+1] = array[j]
        array[j] = key
        j-=1 #Se modifica variable j para seguir recorriendo el arreglo hacia la izquierda hasta llegar a la posición 0
      else:
        break #De no ser necesario una permutación, se sigue avanzando para el siguiente número
    
    if verbose == True:
      print("Partes ordenadas del arreglo luego de permutar:")
      print("[", end = '')

      for k in range(0, i): #Coloreado del arreglo para una mejor comprensión de este
        print("\033[92m" + str(array[k]) + ", ", end = '')
      for l in range(i, len(array)):
        if(l != len(array)-1):
          print("\033[0m" + str(array[l]) + ", ", end = '')
        else:
          print("\033[0m" + str(array[l]), end = '')

      print("]\n")

  return comp, permut

#Ejemplo
A = [6,10,4,7,9]
print("Entrada: ", A)
comp, permut = insertionSort(A)
media = (comp/len(A))
print("Salida: ", A)
print("Comparaciones:", comp)
print("Permutaciones:",  permut)
print("Media de comparaciones:", media)
```

## 2.2 Descripción



El algoritmo de insertion sort consiste en el ordenamiento de un arreglo o lista de números que mediante iteraciones sitúa cada elemento del arreglo en una posición correspondiente.

**Entrada:** Un arreglo o lista con n cantidad de números.

**Salida:** El mismo arreglo con exactamente los mismos elementos pero ordenados de menor a mayor.

El funcionamiento mencionado anteriormente se consigue gracias a estas 3 características del algoritmo:

1. Se comienza a trabajar desde la izquierda hacia la derecha del arreglo, hasta llegar hasta la última posición del arreglo.

2. Cada elemento en una posición i-ésima del arreglo se compara con el número más cercano a su izquierda, y se pregunta si es menor que este, de no ser así, los números son intercambiados.

3. El paso 2 se repite hasta que la comparación se cumpla o hasta que se llegue al incio del arreglo. Se asume que cuando la condición se cumple, todos los números anteriores a la posición en la que queda el nuevo elemento insertado están ya ordenados.

## 2.3 Ejemplo

En el ejemplo del código el arreglo utilizado es el siguiente

* $a = [6,10,4,7,9]$

En la primera pasada el arreglo queda tal cual debido a que el 10 se compara con el 6 y la condición se cumple ya que 6 < 10, por lo cual no hay intercambio.

* $a = [6,10,4,7,9]$

En la segunda pasada el arreglo queda de la siguiente manera:

* $a = [4,6,10,7,9]$ 

Esto es debido a que el número a ubicar es el 4, el cual es menor que los dos números a su izquierda, quedando así en la primera posición del arreglo.

Para la tercera iteración, el arreglo se muestra de la siguiente manera:

* $a = [4,6,7,10,9]$ 

El número a ubicar es el 7, por lo cual intercambia su posición con el 10.

Finalmente, la cuarta y última iteración corresponde a la ubicación del último elemento del arreglo, por lo cual este queda así:

* $a = [4,6,7,9,10]$ 

El 9 es menor que el 10, por lo cual intercambian sus posiciones.

De esta manera es que nos queda el arreglo final con todos sus elementos ordenados de menor a mayor.

* $a = [4,6,7,9,10]$ 




## 2.4 Ejecución del código con opción verbose

```python
import random

A = random.sample(range(1, 100), 6)
print("Input array:", A)
comp, permut = insertionSort(A, verbose = True)
media = (comp/len(A))
print("Salida: ", A)
print("Comparaciones:", comp)
print("Permutaciones:",  permut)
print("Media de comparaciones:", media)
```

Para cada paso se imprime el arreglo antes y después de realizar las modificaciones, el número el cual se quiere posicionar y el estado de los elementos anteriores que ya se encuentran previamente ordenados destacados en verde.

# 3. Propidades del algoritmo

## 3.1Tiempo de ejecución

###**Teorema**

*El tiempo de ejecución del insertion sort es de $O(n)$ en el mejor caso.*

###**Prueba del teorema**

El mejor caso para este algoritmo es que se encuentre ya ordenado, es así como se realizará una cantidad *n-1* comparaciones, siendo *n* el tamaño del arreglo en cuestión. Ya que en primer lugar, nótese que el primer elemento no se compara debido a que se encuentra trivialmente ordenado, luego, por cada paso se irá comparando solo una vez el elemento `A[i]` con el número `A[i-1]` hasta llegar a el último componente del arreglo. Así, aplicando la notación O grande, eliminamos los valores numéricos que acompañan a la variable que crece más rápido y nos da un resultado de O(n) para el tiempo de ejecución.

###**Ejemplo del teorema**



```python
A = [1,2,3,4]
print("Input array:", A)
comp, permut = insertionSort(A)
media = (comp/len(A))
print("Salida: ", A)
print("Comparaciones:", comp)
print("Permutaciones:",  permut)
print("Media de comparaciones:", media)
```

En este caso, el algoritmo nos devuelve que se realizaron 3 comparaciones totales, es decir $n-1$ comparaciones ya que $n = 4$, y por otro lado se realizaron 0 permutaciones debido a que nunca se cumplió la condición de `A[i-1] > A[i]`.

## 3.2 Correctitud

### **Teorema**

El algoritmo insertion sort, luego de ser aplicado en un arreglo del orden $[a_1,a_2, \ldots ,a_n]$, genera un nuevo arreglo $[a_1',a_2', \ldots ,a_n']$ el cual cumple la condición de estar ordenado de **menor a mayor**, es decir $a_1' \leq a_2' \leq \ldots \leq a_n'$ 

### **Prueba del teorema**

El teorema planteado se puede comprobar mediante al siguiente **propiedad invariante de bucle**, y es que cada elemento perteneciente al sub arreglo **[ 0,..., A[i-1] ]**, anterior a cualquier elemento **A[i]**, contiene los mismos elementos del arreglo ordenados de menor a mayor.

**Inicialización**

Para la primera iteración, los elementos anteriores al elemento `A[1]` se encuentran trivialmente ordenados,  ya que el subarreglo mencionado tiene solo 1 elemento (el elemento `A[0]`).

**Mantención**

Para la *i*-ésima iteración, se asume que los números anteriores al elemento `A[i]` están previamente ordenados de menor a mayor. Luego al momento de la inserción, se irá moviendo el elemento `A[i]` hasta que se cumpla que `A[i-1] < A[i]`, de modo que todos los elementos mayores se encontrarán a la derecha del número insertado y los menores a la izquierda, manteniendo el orden  y los componentes para el nuevo arreglo **[ 0,..., A[i] ]**.

Al comprobar que esta propiedad se cumple al inicio del algoritmo y durante la iteración *i-ésima*, se asume que para el último elemento la propiedad **invariante del bucle** se sigue cumpliendo y por ende queda demostrada.

# 4. Experimentos

## 4.1 Mejor y peor caso

Para el primer experimento, se estudiarán la cantidad de comparaciones realizadas en el mejor y peor caso posible para el algoritmo.

**Mejor caso:** Que todos los elementos $a_n$ del arreglo ya se encuentren ordenados de tal manera que $a_1 \leq a_2 \leq \ldots \leq a_n$. Para este caso la cantidad de comparaciones viene dada por $n-1$ como se mencionó anteriormente.

**Peor caso:** Que los elementos se encuentren ordenados de mayor a menor, es decir $a_1 \geq a_2 \geq \ldots \geq a_n$. En este caso, la cantidad de comparaciones viene dada por $n^2 - n \above 1pt 2$.

```python
import matplotlib.pyplot as plt

x=[n for n in range(5,20)] 
y1=[(n*n - n) / 2 for n in range(5,20)] # worst case
y2=[n-1 for n in range(5,20)] # best case
y=[]; 

for n in range(5,20):
  a = random.sample(range(1, 100), n)
  a,counter = insertionSort(a)
  y.append(counter)

plt.plot(x,y1)
plt.plot(x,y2)
plt.legend(["Peor caso", "Mejor caso"])

plt.xlabel('Tamaño del problema')
plt.ylabel('Número de comparaciones')
plt.show()
```

### **Análisis de resultados**

Para la primera gráfica se muestra la relación entre el tamaño $n$ del problema y la cantidad de comparaciones necesaria calculada matemáticamente para el mejor y el peor caso. Se puede observar como la cantidad de comparaciones necesarias para el peor caso crece más rapido que para el mejor, debido a que el primero tiene una complejidad cuadrática mientras que el segundo tiene una complejidad lineal.

## 4.2 Arreglo casi completamente ordenado

Para el siguiente experimento, se estudiará que es lo que sucede para un arreglo ya ordenado, excepto por un dato fuera de posición, se analizará lo que ocurre con la cantidad de comparaciones y permutaciones a medida que los números fuera de posición se encuentran más alejados entre ellos.


```python
A = [5,2,3,4,1]

print("Input array:", A)
comp, permut = insertionSort(A)
media = (comp/len(A))
print("Salida: ", A)
print("Comparaciones:", comp)
print("Permutaciones:",  permut)
print("Media de comparaciones:", media)
```

```python
A = [1,4,3,2,5]

print("Input array:", A)
comp, permut = insertionSort(A)
media = (comp/len(A))
print("Salida: ", A)
print("Comparaciones:", comp)
print("Permutaciones:",  permut)
print("Media de comparaciones:", media)
```

Se puede ver que la cantidad de comparaciones aumenta en la medida que aumente la cantidad de números que hay entre los elementos fuera de posicicón.

En el primer caso, los elementos fuera de posición son el `1` y el `5`, que se encuentran en las posiciones `A[0]` y `A[n-1]` respectivamente, es decir los extremos del arreglo. En esta instancia tenemos 11 comparaciones y 7 permutaciones.

Para el segundo caso, los números desordenados son el `2` y el `4`, que se encuentran en posiciones más cercanas una de otra. De esta forma, la cantidad de comparaciones baja a 7 y las permutaciones necesarias para ordenar bajaron a 3.

Podemos concluir que existe una relación inversamente proporcional entre la separación de los números a ordenar con la cantidad de comparaciones y permutaciones, es decir que entre más separados se encuentren los números a ordenar, la cantidad de operaciones para restaurar el orden va a ser mayor.

```python
import matplotlib.pyplot as plt
import datetime
from timeit import repeat

x=[]; y=[]

for n in range(5,500):

  a = random.sample(range(1, 1000), n)

  #la función repeat está ejecutando 20 veces bubble_sort con el arreglo a y retorna el mejor tiepo obtenido.
  t = repeat(setup="from __main__ import insertionSort", stmt=f"insertionSort({a})", repeat=1, number=10)

  x.append(n)
  y.append(t)


plt.plot(x,y)
plt.legend(["InsertionSort"])

plt.xlabel('n')
plt.ylabel('time in ms')
plt.show()
```

Luego de analizar ambos gráficos, se puede ver que a pesar de ciertos picos de rendimiento, existe un comportamiento parecido entre ambos algoritmos, ya que presentan un crecimiento cuadrático en la relación tamaño del problema-tiempo de ejecución. Nótese también que el algoritmo Insertion Sort tiene un crecimiento un poco más rápido que el bubble sort, lo que implica que se demora un poco más que el Bubble Sort al tratar con problemas muy grandes.
